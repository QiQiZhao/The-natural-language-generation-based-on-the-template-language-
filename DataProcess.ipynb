```python
Data=np.array(Data)
```

```python
for path in csv_path:
    Pmajor=pd.read_csv(path)
    MajorOrderLIst.append(path[0:-4])
    BaseArray=np.empty((len(Data),5),dtype='object')
    for i in range(len(Pmajor)):
        for j in range(len(Data)):
            if Pmajor['Name of Institution'][i]==Data[j][1]:
                BaseArray[j][0]=Pmajor['Rank'][i]
                BaseArray[j][1]=Pmajor['Guardian score/100'][i]
                BaseArray[j][2]=Pmajor['% Satisfied with Teaching'][i]
                BaseArray[j][3]=Pmajor['% Satisfied with course'][i]
                BaseArray[j][4]=Pmajor['Continuation '][i]
                break
    Data=np.hstack((Data,BaseArray))
```
